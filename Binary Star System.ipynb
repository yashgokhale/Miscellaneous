```python
def orbitcalculator(t,PC,m,i,l,e,T,p,arg,m1,m2):
    n=2*3.14/p 
    M=n*(t-T)
    E_latest=optimize.newton(f,0,fprime=lambda x:e*np.cos(x)-1)
    TA=np.arccos((np.cos(E_latest)-e)/(1-e*np.cos(E_latest)))
    Sep_dist_RP=PC*m
    Sep_dist_OP = Sep_dist_RP/(((np.sin((arg+TA)*1.543)**2+(np.cos((arg+TA))**2))**0.5))
    Sep_dist_RP_Y = Sep_dist_OP*np.sin((arg+TA)*1.543)
    Sep_dist_RP_X = Sep_dist_OP*np.cos((arg+TA))
    ang_AN_to_M1_corr=np.arctan(Sep_dist_RP_Y/Sep_dist_RP_X) 
    ang_AN_to_M1_corr_deg=ang_AN_to_M1_corr*360/2/np.pi
    if ang_AN_to_M1_corr_deg < 90:
        Position_Angle=l+ang_AN_to_M1_corr_deg+180
    else:
        Position_Angle=l+ang_AN_to_M1_corr_deg
    a_total=Sep_dist_OP*(1+e*np.cos(TA))/(1-e**2)
    X=m2/m1
    a1=a_total/(1+X)
    a2=a1/X
    d_stars_m=1.496*(10**11)*a_total
    return Position_Angle
```
